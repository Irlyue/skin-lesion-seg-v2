```python
def count_many(predictions, labels):
    """
    Function that return many counter, including TP, TN, FP, FN.
    Labels are assumed to have only two classes, namely 0 and 1.
    :param predictions:
    :param labels:
    :return:
        result: dict
    """
    assert predictions.shape == labels.shape, \
        'shape of predictions%s not equal to labels%s' % (predictions.shape, labels.shape)
    assert (set(np.unique(predictions)) | set(np.unique(labels))) == {0, 1}, \
        '<<ERROR>> classes set not equal {0, 1}'

    correct = (predictions == labels)
    wrong = (predictions != labels)
    true_positives = int(np.sum(correct & (labels == 1)))
    true_negatives = int(np.sum(correct & (labels == 0)))
    false_positives = int(np.sum(wrong & (predictions == 1)))
    false_negatives = int(np.sum(wrong & (predictions == 0)))
    result = {
        'TP': true_positives,
        'TN': true_negatives,
        'FP': false_positives,
        'FN': false_negatives,
    }
    return result


def metric_many_from_counter(result):
    TP, TN, FP, FN = result['TP'], result['TN'], result['FP'], result['FN']
    accuracy = (TP + TN) * 1.0 / (TP + TN + FP + FN)
    sensitivity = TP * 1.0 / (TP + FN)
    specificity = TN * 1.0 / (TN + FP)
    out = {
        'accuracy': accuracy,
        'sensitivity': sensitivity,
        'specificity': specificity
    }
    return out


def metric_many_from_predictions(predictions, labels):
    result = count_many(predictions, labels)
    return metric_many_from_counter(result)
```

```python
data = inputs.load_training_data('dermquest', config)
```

```python
with tf.Graph().as_default():
    global_step = tf.train.get_or_create_global_step()
    image_ph, _, _ = model.model_placeholder(config)
    mm = model.Model(image_ph, config['input_size'])

    saver = tf.train.Saver()
    with tf.Session() as sess:
        utils.load_model(saver, config)
        print('Model-%i restored successfully!' % sess.run(global_step,))
        ops = [mm.endpoints['bbox'], mm.endpoints['lesion_mask'], mm.endpoints['lesion_prob']]
        result = {
            'TP': 0,
            'TN': 0,
            'FP': 0,
            'FN': 0,
        }
        
        def update_dict(target, to_update):
            for key in to_update:
                target[key] += to_update[key]
                
        for i in range(len(data.listing)):
            image, label, bbox = data[i]
            feed_dict = {image_ph: image}
            bbox, lesion_mask, lesion_prob = sess.run(ops, feed_dict)
            bbox[2] = lesion_prob.shape[1]
            bbox[3] = lesion_prob.shape[2]
#             cnn_result = evaluation.get_cnn_result(image.shape[:-1], bbox, lesion_mask)
            cnn_crf_result = evaluation.get_cnn_crf_result(image, bbox, lesion_prob)
            result_i = count_many(cnn_crf_result, label)
            update_dict(result, result_i)
            if i % 10 == 0:
                print(i, ' '.join('{:<15} {:.3f}'.format(key, value) for key, value in 
                                  metric_many_from_counter(result_i).items()))
        result.update(metric_many_from_counter(result))
        print(result)
```
